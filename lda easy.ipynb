```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score

iris=load_iris()
x=iris.data
y=iris.target

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)

print("before: ",x_train.shape)

model=LinearDiscriminantAnalysis(n_components=2)
x_lda=model.fit_transform(x_train,y_train)

print("after lda: ",x_lda.shape)

pred=model.predict(x_test)
accuracy=accuracy_score(pred,y_test)
print("accuracy: ",accuracy*100)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

iris=load_iris()

x=iris.data
y=iris.target

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

print("before: ",x_train.shape)

lda=LinearDiscriminantAnalysis(n_components=2)
x_lda=lda.fit_transform(x_test,y_test)

knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_lda,y_test)

print("after: ",x_lda.shape)

pred=knn.predict(x_lda)
accu=accuracy_score(pred,y_test)
print("accuary: ",accu)
print(classification_report(pred,y_test,target_names=iris.target_names))
```

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import load_iris

iris=load_iris()
x=iris.data
y=iris.target

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

print("before: ",x_train.shape)

lda=LinearDiscriminantAnalysis(n_components=2)
x_lda=lda.fit_transform(x_test,y_test)

print("after: ",x_lda.shape)

kmeans=KMeans(n_clusters=3,n_init=10)
x_kmeans=kmeans.fit(x_lda,y_test)

pred=kmeans.predict(x_lda)
acc=accuracy_score(y_test,pred)
print("acc: ", acc*100)
```
