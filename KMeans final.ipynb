```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Choose the number of clusters
n_clusters = 3

# Create a KMeans model
model = KMeans(n_clusters=n_clusters, random_state=42)

# Fit the model to the data
model.fit(X)

# Get cluster assignments for each data point
cluster_assignments = model.labels_

# Visualize the clusters in 2D (using the first two features)
plt.scatter(X[:, 0], X[:, 1], c=cluster_assignments, cmap='coolwarm')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], c='yellow', marker='X', s=400, label='Cluster Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('K-Means Clustering of Iris Dataset')
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Choose the number of clusters
n_clusters = 3

# Create a KMeans model
model = KMeans(n_clusters=n_clusters, random_state=42)

# Split the data into a train and test set (for evaluation)
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

# Fit the model to the training data
model.fit(X_train)

# Get cluster assignments for each data point in the training set
train_cluster_assignments = model.predict(X_train)

# Visualize the clusters in 2D (using the first two features) for the training set
plt.scatter(X_train[:, 0], X_train[:, 1], c=train_cluster_assignments, cmap='coolwarm')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], c='yellow', marker='X', s=400, label='Cluster Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('K-Means Clustering of Iris Dataset (Training Set)')
plt.show()

# Get cluster assignments for each data point in the test set
test_cluster_assignments = model.predict(X_test)

# Visualize the clusters in 2D (using the first two features) for the test set
plt.scatter(X_test[:, 0], X_test[:, 1], c=test_cluster_assignments, cmap='coolwarm')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], c='yellow', marker='X', s=400, label='Cluster Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('K-Means Clustering of Iris Dataset (Test Set)')
plt.show()
```
