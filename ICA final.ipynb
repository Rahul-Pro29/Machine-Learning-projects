```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import FastICA

# Generate two source signals
np.random.seed(0)
n_samples = 200
time = np.linspace(0, 10, n_samples)

# Source signal 1: A sinusoidal signal
s1 = np.sin(2 * time)

# Source signal 2: A square wave signal
s2 = np.sign(np.sin(3 * time))

# Combine the source signals to create the observed mixed signals
S = np.c_[s1, s2]
A = np.array([[1, 1], [0.5, 2]])  # Mixing matrix
X = np.dot(S, A.T)

# Apply ICA to separate the mixed signals
ica = FastICA(n_components=2)
S_ = ica.fit_transform(X)

# Plot the original and separated signals
plt.figure(figsize=(10, 6))

# Original source signals
plt.subplot(3, 1, 1)
plt.title("Original Source Signals")
plt.plot(S)

# Mixed signals
plt.subplot(3, 1, 2)
plt.title("Mixed Signals")
plt.plot(X)

# Separated signals using ICA
plt.subplot(3, 1, 3)
plt.title("Separated Signals (ICA)")
plt.plot(S_)

plt.tight_layout()
plt.show()
```
