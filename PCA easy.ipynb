knn

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score,classification_report

iris=load_iris()
x=iris.data
y=iris.target

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

print("before:" ,x_train.shape)

pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_test,y_test)

print("after: ",x_pca.shape)

knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_pca,y_test)

pred=knn.predict(x_pca)
acc=accuracy_score(pred,y_test)

print("acc: ",acc*100)
print(classification_report(pred,y_test,target_names=iris.target_names))
```

kmeans

```python
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

iris=load_iris()
x=iris.data
y=iris.target

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

print("before: ",x_train.shape)

pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_test,y_test)

print("after: ",x_pca.shape)

kmeans=KMeans(n_clusters=3,n_init=10)
kmeans.fit(x_pca,y_test)

pred=kmeans.predict(x_pca)
acc=accuracy_score(pred,y_test)
print("acc: ",acc*100)

```
