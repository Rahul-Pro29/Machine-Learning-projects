## **OR gate**

```python
import numpy, random, os
lr = 0.5 #learning rate
bias = 0.5 #value of bias
weights = [0.9,0.9,0.9] #weights generated in a list (3 weights in total for 2 neurons and the bias)
def Perceptron(input1, input2, output) :
   outputP = input1*weights[0]+input2*weights[1]+bias*weights[2]
   if outputP > 0 : #activation function (here Heaviside)
      outputP = 1
   else :
      outputP = 0
   error = output - outputP
   weights[0] += error * input1 * lr
   weights[1] += error * input2 * lr
   weights[2] += error * bias * lr
for i in range(50) :
   Perceptron(1,1,1) #True or true
   Perceptron(1,0,1) #True or false
   Perceptron(0,1,1) #False or true
   Perceptron(0,0,0) #False or false
x = int(input())
y = int(input())
outputP = x*weights[0] + y*weights[1] + bias*weights[2]
if outputP > 0 : #activation function
   outputP = 1
else :
   outputP = 0
print(x, "or", y, "is : ", outputP)
outputP = 1/(1+numpy.exp(-outputP)) #sigmoid function
print(outputP)
```

# **NAND gate**


```python
import numpy
lr = 0.5  # Learning rate
bias = 0.5  # Value of bias
weights = [-0.9, -0.9, -0.9]  # Weights for the NAND gate (opposite of OR gate)

def Perceptron(input1, input2, output):
    outputP = input1 * weights[0] + input2 * weights[1] + bias * weights[2]
    if outputP > 0:  # Activation function (Heaviside)
        outputP = 1
    else:
        outputP = 0
    error = output - outputP
    weights[0] += error * input1 * lr
    weights[1] += error * input2 * lr
    weights[2] += error * bias * lr

# Training the NAND gate
for i in range(50):
    Perceptron(1, 1, 0)  # True and true (opposite of OR gate)
    Perceptron(1, 0, 1)  # True and false (opposite of OR gate)
    Perceptron(0, 1, 1)  # False and true (opposite of OR gate)
    Perceptron(0, 0, 1)  # False and false (opposite of OR gate)

x = int(input())
y = int(input())
outputP = x * weights[0] + y * weights[1] + bias * weights[2]

if outputP > 0:  # Activation function
    outputP = 1
else:
    outputP = 0

print(x, "nand", y, "is:", outputP)
outputP = 1 / (1 + numpy.exp(-outputP))  # Sigmoid function
print(outputP)
```

## **AND gate**

```python
import numpy

lr = 0.5  # Learning rate
bias = -0.5  # Value of bias
weights = [0.9, 0.9, 0.9]  # Weights for the AND gate

def Perceptron(input1, input2, output):
    outputP = input1 * weights[0] + input2 * weights[1] + bias * weights[2]
    if outputP > 0:  # Activation function (Heaviside)
        outputP = 1
    else:
        outputP = 0
    error = output - outputP
    weights[0] += error * input1 * lr
    weights[1] += error * input2 * lr
    weights[2] += error * bias * lr

# Training the AND gate
for i in range(50):
    Perceptron(1, 1, 1)  # True and true
    Perceptron(1, 0, 0)  # True and false
    Perceptron(0, 1, 0)  # False and true
    Perceptron(0, 0, 0)  # False and false

x = int(input())
y = int(input())
outputP = x * weights[0] + y * weights[1] + bias * weights[2]

if outputP > 0:  # Activation function
    outputP = 1
else:
    outputP = 0

print(x, "and", y, "is:", outputP)
outputP = 1 / (1 + numpy.exp(-outputP))  # Sigmoid function
print(outputP)
```
